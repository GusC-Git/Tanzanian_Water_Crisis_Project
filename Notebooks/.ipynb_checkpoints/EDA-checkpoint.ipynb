```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
os.getcwd()
```

```python
df = pd.read_csv("../Data/scrubbed_trainingdata.csv")
df.drop(columns= "Unnamed: 0",inplace=True)
display(df.head())
df.info()
```

## Q1: What is the class balance?

```python
#Creating a custom palette for 
colors = ['#04F333','#F32804','#E5E813']
custompalette = sns.set_palette(sns.color_palette(colors))
```

```python
sns.set_style("dark")
```

```python
import matplotlib.ticker as ticker
plt.figure(figsize=(12,8))
cb = sns.countplot(x= 'status_group', data=df, hue= 'status_group',dodge=False, \
              hue_order= ['functional','non functional','functional needs repair'], palette= custompalette)
#Makes a twin axis
cb2 = cb.twinx()

#Sets frequency on the left and adjusts label
cb2.yaxis.tick_left()
cb2.yaxis.set_label_position('left')
cb2.set_ylabel('Percentage of Total Pumps [%]', fontsize=14)

#Sets count on the right and adjusts label
cb.yaxis.tick_right()
cb.yaxis.set_label_position('right')

#Setting labels
cb.set_xlabel("Status Group", fontsize=14)
cb.set_ylabel("Count", fontsize=14)
cb.set_title("Value Counts of Pump Status", fontsize=18)
for patch in cb.patches:
    x = patch.get_bbox().get_points()[:,0]
    y = patch.get_bbox().get_points()[1,1]
    cb.annotate("{:.1f}%".format(100.*y/len(df)), (x.mean(), y),\
               ha= 'center', va= 'bottom', fontsize= 'large') #sets text alignment

cb.yaxis.set_major_locator(ticker.LinearLocator(10))

cb2.set_ylim(1,100)
cb.set_ylim(0, len(df))



plt.show()
```

There is significantly more functional and non functional pumps than there are functional that need repair. I will have to address this during the modeling. If the data is too imbalanced, the model could miss classification of the minority class almost entirely.

## Q2. What is the distribution of the functionality in each basin?

```python
plt.figure(figsize=(12,8))
ct = sns.countplot(y='basin', data= df, hue = 'status_group', hue_order= df['status_group'].value_counts().index,\
             order = df['basin'].value_counts().index, palette=custompalette)

ct.set_xlabel("Number of Pumps", fontsize=14)
ct.set_ylabel("Basin", fontsize=14)
ct.set_title("Count and Pump Status per Basin", fontsize=16)

ct.legend(title='Status Group')
plt.show()
```

There is a good spread of pumps across Tanzania but we can see that functionality of pump varies per basin. While Lake Victoria basin outnumbers the others in total number of pumps, they have more non functional pumps and less functional pumps than Pangani. From this plot it can also be identified that certain basins like Ruvuma/Southern Coast basin and Lake Rukwa are under-served, both basins having a larger number of non functional pumps than they do functional, and also being the two basins with the least number of pumps.

## Q3 what is the population in each basin?

```python
#Creates cut of dataframe to get population and basin
basin_pop_df = df[['basin','population']].groupby(by='basin').sum().reset_index()

plt.figure(figsize=(12,8))
#Plots the population located in each basin
pb = sns.barplot(data= basin_pop_df, y='basin', x='population', orient='h',order = df['basin'].value_counts().index,\
                 palette=sns.color_palette("tab10"))


pb.set_ylabel('Basin', fontsize=14)
pb.set_xlabel('Population', fontsize=14)
pb.set_title('Population per Basin', fontsize=16)
plt.show()
```

When this bar graph is cross-referenced with the graph of the previous question, we see that many people live in the Ruvuma / Southern Coast area, yet it is one of two basins who have more non funcitonal pumps than functional as well as one of the two basins with the least number of pumps. This could be indicative that the community there is under-served and could be an important target for focused efforts when it comes to repairing the pumps.

## Q4. What is the relationship between Source type and  functionality?


```python
plt.figure(figsize=(15,8))
sc = sns.countplot(x='source_type', data= df, hue = 'status_group',\
                   hue_order= df['status_group'].value_counts().index,\
                  order = df['source_type'].value_counts().index, palette=custompalette)

sc2 = sc.twinx()

#Sets frequency on the left and adjusts label
sc2.yaxis.tick_left()
sc2.yaxis.set_label_position('left')
sc2.set_ylabel('Percentage of Total Pumps [%]', fontsize=14)

#Sets count on the right and adjusts label
sc.yaxis.tick_right()
sc.yaxis.set_label_position('right')

for patch in sc.patches:
    x = patch.get_bbox().get_points()[:,0]
    y = patch.get_bbox().get_points()[1,1]
    sc.annotate("{:.1f}%".format(100.*y/len(df)), (x.mean(), y),\
               ha= 'center', va= 'bottom') #sets text alignment

sc.yaxis.set_major_locator(ticker.LinearLocator(12))

sc2.set_ylim(1,100*round((len(df)+1000)/3, -3)/round((len(df)+1000), -3))
sc.set_ylim(0, round((len(df)+1000)/3, -3))

sc.set_xlabel("Source Type", fontsize=14)
sc.set_ylabel("Number of Pumps", fontsize=14)
sc.set_title("Pump Functionality per Source Type", fontsize=16)

plt.show()
```

Shallow well and borehole sourced water pumps about evenly match in ratios of functioning pumps to non functioning/need repair pumps. Since both are groundwater methods, it appears that groundwater sources are liable to have less functioning pumps in relation to the number non functioning pumps. Very little water is gathered through other means such as dams, rainwater catchement. It also appears that groundwater sources such as wells and boreholes are less likely to have wells that function but need repair.

## Q5. What is the relationship between source type and  water quantity?

```python
plt.figure(figsize=(15,8))
sw = sns.countplot(x='source_type', data= df, hue = 'quantity',\
                   hue_order= df['quantity'].value_counts().index,\
                  order = df['source_type'].value_counts().index, palette=sns.color_palette("viridis"))

sw.set_xlabel("Source Type", fontsize=14)
sw.set_ylabel("Number of Pumps", fontsize=14)
sw.set_title("Water Quantity per Source Type", fontsize=16)
plt.show()
```

Rainwater harvesting appears to be the most seasonal of all quantity amounts. Spring water is the most common source of water as well as the source with the most number of pumps that provide enough water. Borehole and Shallow well pumps have a higher chance of being dry than pumps that get their water from a spring or river/lake.

```python
plt.figure(figsize=(15,8))
sq = sns.countplot(x='quantity', data= df, hue = 'status_group',\
                   hue_order= df['status_group'].value_counts().index,\
                  order = df['quantity'].value_counts().index, palette=custompalette)

sq.set_xlabel("Quantity Amount", fontsize=14)
sq.set_ylabel("Number of Pumps", fontsize=14)
sq.set_title("Pump Functionality per Water Quantity", fontsize=16)

plt.show()
```

The majority of the pumps that are able to provide enough water are also functional. Almost every pump that has now gone dry is classified as non functional. This gives insight into why certain pumps, specifically these, could possible not be working--if a pump has gone dry, it can be considered non functional then. A priority here would be returning to functional and giving the needed repairs to the pumps that provide enough water.

## Q6. How does extraction type relate to functionality and Water Quantity?


```python
plt.figure(figsize=(15,8))

ec = sns.countplot(y='extraction_type_class', data= df, hue = 'status_group',\
                   hue_order= df['status_group'].value_counts().index,\
                  order = df['extraction_type_class'].value_counts().index, palette= custompalette)

ec.set_xlabel("Number of Pumps", fontsize=14)
ec.set_ylabel(" Water Extraction Type", fontsize=14)
ec.set_title("Pump Functionalbility by Extraction Type", fontsize=16)

plt.show()
```

Gravity is the preferred method of extraction for most pumps and has the highest number of functional pumps. Motorpumps also have a higher likelihood of being non functional than functional. Whenever a pump is classified as 'other' there is a good chance that it is nonfunctional and was classified as such because of that. Gravity pumps also have the highest number of functional but in need of repairs pumps. Due to their plentitude, a good course of action would be to address the issues in gravity pumps that have it and fix it. the majority of pumps that are functional and need repair are gravity pumps.

```python
plt.figure(figsize=(15,8))
eeq = sns.countplot(y='extraction_type_class', data= df, hue = 'quantity',\
                   hue_order= df['quantity'].value_counts().index,\
                  order = df['extraction_type_class'].value_counts().index, palette= sns.color_palette("viridis"))

eeq.set_ylabel("Water Extraction Type", fontsize=14)
eeq.set_xlabel("Number of Pumps", fontsize=14)
eeq.set_title("Water Quantity per Extraction type", fontsize=16)

eeq.legend(loc='lower right')
plt.show()
```

In relaiton to the previous graph, here we see that gravity pumps are also more often than not providing providing more water than any other type of pump. It can be assumed that for both gravity and handpump, all functioning wells are also providing enough water, since the numbers match up. 

 Interesting is that the vast majority of pumps if they are of spring water will be a gravity type of pump and spring is also the most common water source.
